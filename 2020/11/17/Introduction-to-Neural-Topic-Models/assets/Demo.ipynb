```python
import numpy as np 
import umap
from sklearn.manifold import TSNE
```

```python
import matplotlib.pyplot as plt 
```

```python
z_dim = 10
batch_size = 500
odes = np.eye(z_dim)*20
ides = np.random.randint(low=0, high=z_dim, size=batch_size)
mus = odes[ides]
sigmas = np.ones((batch_size, z_dim))*0.2*20
z_true = np.random.normal(mus, sigmas)
spls = z_true
#z_true = F.softmax(torch.from_numpy(z_true).float(), dim=1)
```

```python
n = 5
spls = np.random.dirichlet(alpha=np.ones(n)/n,size=500)
```

```python
red_emb = umap.UMAP(n_components=2).fit_transform(spls)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(red_emb[:,0],red_emb[:,1],s=10,c='darkorange')
plt.savefig('dirichlet.png')
plt.show()
```

```python
red_emb_t = TSNE(n_components=2).fit_transform(spls)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(red_emb_t[:,0],red_emb_t[:,1],s=10,c='darkorange')
plt.savefig('dirichlet.png')
plt.show()
```
